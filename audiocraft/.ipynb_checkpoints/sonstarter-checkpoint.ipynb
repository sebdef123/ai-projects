```python
from audiocraft.models import musicgen

model = musicgen.MusicGen.get_pretrained('nateraw/musicgen-songstarter-v0.1')
```

```python
from datetime import datetime
from pathlib import Path


from audiocraft.data.audio import audio_write
from audiocraft.utils.notebook import display_audio


# path to save our samples.
out_dir = Path("./samples")
out_dir.mkdir(exist_ok=True, parents=True)

model.set_generation_params(
    duration=2,
    use_sampling=True,
    temperature=1.0,
    top_k=250,
    cfg_coef=3.0,
)

text = "hip hop, soul, piano, chords, jazz, neo jazz, G# minor, 140 bpm"
N = 4
out = model.generate(
    [text] * N,
    progress=True,
)

# Write to files
dt_str = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
for i in range(N):
    audio_write(
        out_dir / f"{dt_str}_{i:02d}",
        out[i].cpu(),
        model.sample_rate,
        strategy="loudness",
    )

# Or, if in a notebook, display audio widgets
# display_audio(out, model.sample_rate)
```

```python
display_audio(out, model.sample_rate)
```
