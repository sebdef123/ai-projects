## Step 1: Download images of birds and non-birds

```python

def search_images(term, max_images=100):
    print(f"Searching for '{term}'")
    return L(search_images_ddg(term, max_images=max_images))
```

Let's start by searching for a bird photo and seeing what kind of result we get. We'll start by getting URLs from a search:

```python
import os
root_dir = os.getcwd()
root_dir='../../'
images_dir=root_dir+'dataset/images'
path = Path(images_dir)
path.mkdir(exist_ok=True)
```

```python
searches = 'doublebass','drumkit','saxophone','clarinet'

for search in searches:
    dest = (path/f'{search}')
    print (dest)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{search} photo', max_images=90))

    resize_images(path/f'{search}', max_size=400, dest=path/f'{search}')
```

```python
search = 'guitar'

dest = (path/'guitar')
dest.mkdir(exist_ok=True, parents=True)
download_images(dest, urls=search_images(f'{search} photo', max_images=30))
sleep(10)  # Pause between searches to avoid over-loading server

download_images(dest, urls=search_images(f'{search} vertical photo', max_images=30))
sleep(10)
download_images(dest, urls=search_images(f'{search} horizontal photo', max_images=30))
sleep(10)

resize_images(path/'guitar', max_size=400, dest=path/'guitar')
```
